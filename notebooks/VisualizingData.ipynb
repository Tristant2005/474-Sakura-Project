```python
import pandas as pd
```

```python
df = pd.read_csv("/content/all_cities_cat_v2.csv")
```

```python
print(df.columns)
display(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_1981 = df[df['seasonal_year'] == 1981]

plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=df_1981,
    x='longitude',
    y='latitude',
    s=100, # Constant size for points
    hue='Mar_temp',
    palette='viridis',
    # legend='full'
)
plt.title('Scatter Plot of Cities by Latitude, Longitude, and March 1981 Temperature')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.grid(True)
plt.show()
```
